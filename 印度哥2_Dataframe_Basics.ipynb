#基本操作與數學運算

```python
import pandas as pd
# df=pd.read_csv('/content/weather_data.csv')
# df
```

```python
#透過字典來建立Dataframe
weather_data = {
    'day': ['1/1/2017','1/2/2017','1/3/2017','1/4/2017','1/5/2017','1/6/2017'],
    'temperature': [32,35,28,24,32,31],
    'windspeed': [6,7,2,7,4,2],
    'event': ['Rain', 'Sunny', 'Snow','Snow','Rain', 'Sunny']
}
df=pd.DataFrame(weather_data)
df
```

```python
df.shape #shape means dimansion

rows,columns = df.shape #回傳值
print(rows)
print(columns)
```

```python
#head()用來指定only印出想要的前幾個rows數量
df.head(2)
```

```python
#tail()用來指定only印出想要的最後幾個rows數量
df.tail(2)
```

```python
#sclicing用來指定想印出的rows範圍
df[2:5]#sclicing ，從2~不包括5
```

```python
#用來了解你一共有哪些columns
df.columns
```

```python
#選取你想要的特定column印出
print(df.day)
print(df.temperature)
print(df['windspeed']) #也可以用這種語法
```

```python
type(df['event'])
```

```python
#將想要印出的表格欄集中到一張圖表中
df[['day','windspeed','event']]
```

```python
#數學計算
z=df['temperature'].max()
z1=df.temperature.mean() #平均值
z2=df['temperature'].min()
z3=df.temperature.std() #標準差

print(z)
print(z1)
print(z2)
print(z3)
```

```python
#describe()
df.describe()
```

```python
#取出符合特定條件的rows
df[df.temperature>=32]#得到所有溫度>=32的rows
df[df.temperature==df.temperature.max()]#得到temperature為最大值得row
df[df.temperature==df['temperature'].max()] #也能寫成這樣
```

```python
#取出符合特定條件的row，並指定想要印出的column
# df['day','temperature'][df.temperature==df['temperature'].max()] #記住這一行有小錯誤!注意了
df[['day','temperature']][df.temperature==df['temperature'].max()] #這樣才對
```
