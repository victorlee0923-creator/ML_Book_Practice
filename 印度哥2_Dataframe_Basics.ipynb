```python
import pandas as pd
# df=pd.read_csv('/content/weather_data.csv')
# df
```

```python
#透過字典來建立Dataframe
weather_data = {
    'day': ['1/1/2017','1/2/2017','1/3/2017','1/4/2017','1/5/2017','1/6/2017'],
    'temperature': [32,35,28,24,32,31],
    'windspeed': [6,7,2,7,4,2],
    'event': ['Rain', 'Sunny', 'Snow','Snow','Rain', 'Sunny']
}
df=pd.DataFrame(weather_data)
df
```

```python
df.shape #shape means dimansion

rows,columns = df.shape #回傳值
print(rows)
print(columns)
```

```python
#head()用來指定only印出想要的前幾個rows數量
df.head(2)
```

```python
#tail()用來指定only印出想要的最後幾個rows數量
df.tail(2)
```

```python
#sclicing用來指定想印出的rows範圍
df[2:5]#sclicing ，從2~不包括5
```

```python
#選取你想要的特定column印出
print(df.day)
print(df.temperature)
print(df['windspeed']) #也可以用這種語法
```

```python
#數學計算
z=df['temperature'].max()
z1=df.temperature.mean() #平均值
z2=df['temperature'].min()
z3=df.temperature.std() #標準差

print(z)
print(z1)
print(z2)
print(z3)
```
